```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr
%matplotlib inline

cmapl = sns.cubehelix_palette(as_cmap=True)
sns.set(style="darkgrid")
```

```python
data = pd.read_csv("March 7-April 4, 2016 - Libraries - CSV.csv")

readers = pd.DataFrame(data)

focus_df = readers[['live1', 'inc', 'sex', 'educ2', 'books1', 'books2a']]
focus_df.corr()
sns.heatmap(focus_df.corr())
plt.show()

norms = readers[readers['books1']<=20]
nerds = readers[readers['books1']>20]
```

I thought it would be interesting to look at the data surrounding people who read more versus people who read less. As such, I isolated the data into two data frames (nerd/norm).

```python
nerd_data = pd.DataFrame(nerds)
nerd_data
```

```python
norm_data = pd.DataFrame(norms)
norm_data
```

I was curious to see if there was any difference between urban/rural environments and readership (1=city, 2=suburban, 3=small city and 4=rural). The scatterplots below show that there isn't much of a difference. The histograms afterwards, however, show that there is a lot of readership in #3 small cities.

```python
plt.figure(figsize=(10,6));

plt.subplot(1, 2, 1)
plt.hist(nerd_data['books1'].values)
plt.ylabel("Number of Respondents")
plt.xlabel("Number of Books")
plt.title("More than 20 Books/Year")


plt.subplot(1, 2, 2)
plt.hist(norm_data['books1'].values)
plt.ylabel("Number of Respondents")
plt.xlabel("Number of Books")
plt.title("Less than or Equal to 20 Books/Year")

plt.tight_layout()
plt.show()
```

```python
a = nerd_data[nerd_data['live1']<5]
b = a[['books1', 'live1']]
c = norm_data[norm_data['live1']<5]
d = c[['books1', 'live1']]

b.groupby('live1').agg('count')
d.groupby('live1').agg('count')
```

```python
sns.jointplot(x = d['live1'], y = d['books1'], kind="scatter", size = 5, s = 7).set_axis_labels("# of Respondents", "Location")

sns.jointplot(x = d['live1'], y = d['books1'], kind="scatter", size = 5, s = 7).set_axis_labels("# of Respondents", "Location")

plt.show()
```

```python
nerd_data.groupby('sex').agg('count').loc[:, 'books1'].plot(kind="pie", y='sex')


plt.show()
```

```python
norm_data.groupby('sex').agg('count').loc[:, 'books1'].plot(kind='pie', y ='sex')

plt.show()
```
