 I tried to group a few common positive/negative words into some regex expressions - although this is an amazingly more complicated problem than it seems at first.  I was only able to get a small percentage of accurate predictions, but it was a majority that was correct.  Based on the models and outcomes below, it looks like this model gets the best results with Yelp.

```python
# Applying the model above to the different files/contexts:

amazon['positive_words'] = text.Review.str.contains('I lov*|I lik*|it was great*|', case = False, regex = True)
amazon['negative_words'] = text.Review.str.contains('hate*|terribl*|wors*|bad|but|lame|waste*|never again|too',
                                                  case = False, regex = True)

yelp['positive_words'] = text.Review.str.contains('I lov*|I lik*|it was great*|', case = False, regex = True)
yelp['negative_words'] = text.Review.str.contains('hate*|terribl*|wors*|bad|but|lame|waste*|never again|too',
                                                  case = False, regex = True)
```

```python
amazon_positive = amazon.where((amazon['positive_words'] == True) & (amazon['Rating'] == 1)).dropna()
amazon_negative = amazon.where((amazon['positive_words'] == True) & (amazon['Rating'] == 0)).dropna()
```

```python
yelp_positive = yelp.where((yelp['positive_words'] == True) & (yelp['Rating'] == 1)).dropna()
yelp_negative = yelp.where((yelp['positive_words'] == True) & (yelp['Rating'] == 0)).dropna()
```

```python
print("Correctly labelled Amazon reviews: Correct Positives {}, False Positivies {}".format(
    amazon_positive.count(), amazon_negative.count()))
```

```python
print("Correctly labelled Yelp reviews: Correct Positives {}, False Positivies {}".format(
    yelp_positive.count(), yelp_negative.count()))
```
